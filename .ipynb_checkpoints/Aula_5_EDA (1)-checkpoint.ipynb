#### Carga dos dados

```python
ocorrencia = pd.read_csv('http://sistema.cenipa.aer.mil.br/cenipa/media/opendata/ocorrencia.csv',
                         delimiter=';')
```

```python
tipo = pd.read_csv('http://sistema.cenipa.aer.mil.br/cenipa/media/opendata/ocorrencia_tipo.csv',
                   delimiter=';')
```

```python
aeronave = pd.read_csv('http://sistema.cenipa.aer.mil.br/cenipa/media/opendata/aeronave.csv',
                       delimiter=';')
```

```python
fator = pd.read_csv('http://sistema.cenipa.aer.mil.br/cenipa/media/opendata/fator_contribuinte.csv',
                    delimiter=';')
```

```python
recomenda = pd.read_csv('http://sistema.cenipa.aer.mil.br/cenipa/media/opendata/recomendacao.csv',
                        delimiter=';')
```

```python
ocorrencia.head(2)
```

```python
tipo.head(2)
```

```python
aeronave.head(2)
```

```python
recomenda.head(2)
```

```python
fator.head(2)
```

#### Manipulação dos dados

```python
fator.columns
```

```python
fator.head(2)
```

```python
col_ocorrencia = [
    'codigo_ocorrencia', 'codigo_ocorrencia1', 'codigo_ocorrencia2',
    'codigo_ocorrencia3', 'codigo_ocorrencia4', 'ocorrencia_classificacao',
    'ocorrencia_cidade','ocorrencia_uf', 'ocorrencia_aerodromo',
    'ocorrencia_dia', 'total_aeronaves_envolvidas','ocorrencia_saida_pista'
]
col_tipo = [
    'codigo_ocorrencia1', 'ocorrencia_tipo', 'ocorrencia_tipo_categoria'
]
col_aeronave = [
    'codigo_ocorrencia2', 'aeronave_tipo_veiculo', 'aeronave_fabricante', 
    'aeronave_motor_tipo', 'aeronave_motor_quantidade', 'aeronave_registro_segmento',
    'aeronave_fase_operacao','aeronave_tipo_operacao', 'aeronave_nivel_dano'
]
col_recomenda = [
    'codigo_ocorrencia4', 'recomendacao_destinatario_sigla'
]
col_fator = [
    'codigo_ocorrencia3', 'fator_nome', 'fator_aspecto',
    'fator_condicionante', 'fator_area'
]
```

```python
base = ocorrencia[col_ocorrencia]\
          .merge(tipo[col_tipo], how='left', on='codigo_ocorrencia1')\
          .merge(aeronave[col_aeronave], how='left', on='codigo_ocorrencia2')\
          .merge(fator[col_fator], how='left', on='codigo_ocorrencia3')\
          .merge(recomenda[col_recomenda], how='left', on='codigo_ocorrencia4')      
```

```python
base.head(2)
```

```python
base.isna().sum()
```

```python
base = ocorrencia[col_ocorrencia]\
          .merge(tipo[col_tipo], how='left', on='codigo_ocorrencia1')\
          .merge(aeronave[col_aeronave], how='left', on='codigo_ocorrencia2')
```

```python
base.shape
```

```python
base.isnull().sum()
```

#### Limpeza dos dados

```python
base.dropna(inplace=True)
```

```python
base.shape
```

```python
del_col = ['codigo_ocorrencia1','codigo_ocorrencia2',
           'codigo_ocorrencia3','codigo_ocorrencia4']

base.drop(columns=del_col, inplace=True)
```

```python
base.shape
```

```python
base.info()
```

```python
base.head(2)
```

#### Transformando dados

```python
base['ocorrencia_data'] = pd.to_datetime(base['ocorrencia_dia'])
```

```python
print('Menor data:', base['ocorrencia_data'].min())
print('Maior data:', base['ocorrencia_data'].max())
```

```python
base['ocorrencia_dia'] = base['ocorrencia_data'].dt.day
base['ocorrencia_mes'] = base['ocorrencia_data'].dt.month
base['ocorrencia_ano'] = base['ocorrencia_data'].dt.year
```

```python
base.head(2)
```

#### Respondendo a problemas do negócio!!!

##### Qual os 10 estados com mais ocorrencias?

```python
px.bar(data_frame=base, x='ocorrencia_uf')\
  .update_layout(template='presentation')
```

```python
x = base['ocorrencia_uf'].value_counts().head(10).index
y = base['ocorrencia_uf'].value_counts().head(10).values
px.bar(x=x,y=y,text=y)\
  .update_layout(template='presentation')\
  .update_xaxes(title='')\
  .update_yaxes(visible=False)
```

```python
def top_n_col (col,n):
  x = base[col].value_counts().head(n).index
  y = base[col].value_counts().head(n).values
  fig = px.bar(x=x,y=y,text=y)\
        .update_layout(template='presentation')\
        .update_xaxes(title='')\
        .update_yaxes(visible=False)
  return fig
```

```python
top_n_col ('ocorrencia_uf',10)
```

#### Respondendo a problemas do negócio!!!

##### Qual a quantidade de registros por UF?

```python
def top_n_col (df,col,n):
  x = df[col].value_counts().head(n).index
  y = df[col].value_counts().head(n).values
  fig = px.bar(x=x,y=y,text=y)\
        .update_layout(template='presentation')\
        .update_xaxes(title='')\
        .update_yaxes(visible=False)
  return fig
```

```python
top_n_col (df,'UF',20)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
```

```python
#CARREGANDO BASE
df = pd.read_csv('https://raw.githubusercontent.com/pedroprattes/aula-05/main/mcdonalds_dataset.csv')
df.head(10)
```

```python
#LIMPEZA DA BASE
df.isnull().sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['city'].value_counts()
```

```python
#EQUIPAMENTOS QUEBRADOS X EQUIPAMENTOS EM FUNCIONAMENTO
g1 = px.pie(data_frame=df, names='is_broken')

g1.update_layout(
    legend_title='Em funcionamento: "True", Quebrado: "False"',
    title={
        'text': 'Equipamentos em funcionamento x Quebrados',
        'x': 0.5,
        'y': 0.93
    },
    template='simple_white'
)
```

```python
#EQUIPAMENTOS INATIVOS X EQUIPAMENTOS INATIVOS
g2 = px.pie(data_frame=df, names='is_active')

g2.update_layout(
    legend_title='Ativos: "True", Inativos: "False"',
    title={
        'text': 'Equipamentos Ativos x Inativos',
        'x': 0.5,
        'y': 0.93
    },
    template='simple_white'
)
```

```python
df['dot'].value_counts()
```

```python
#EQUIPAMENTOS ATIVOS x SITUAÇÃO ATUAL

sns.countplot(data=df, x='is_broken', hue='is_active');
```

```python
df_filtered = df[df['is_broken'] == True]
df_filtered.shape
```

```python
#TOP 5 ESTATOS - EQUIPAMENTOS QUEBRADOS

df2 = df[df['is_broken'] == True]

def top_n_col (col,n):
  x = df2[col].value_counts().head(n).index
  y = df2[col].value_counts().head(n).values
  fig = px.bar(x=x,y=y,text=y)\
        .update_layout(template='presentation')\
        .update_xaxes(title='')\
        .update_yaxes(visible=False)
  return fig
top_n_col ('state',5)
```

```python
#TOP 5 ESTATOS - EQUIPAMENTOS INATIVOS

df3 = df[df['is_active'] == False]

def top_n_col (col,n):
  x = df3[col].value_counts().head(n).index
  y = df3[col].value_counts().head(n).values
  fig = px.bar(x=x,y=y,text=y)\
        .update_layout(template='presentation')\
        .update_xaxes(title='')\
        .update_yaxes(visible=False)
  return fig
top_n_col ('state',5)
```
